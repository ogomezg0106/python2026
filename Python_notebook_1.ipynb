#### Jerarquía de Operaciones (Regla PEMDAS)

En Python, cuando tenemos varias operaciones en una sola línea, el lenguaje sigue un orden de prioridad específico conocido como **PEMDAS**.  
Si dos operaciones tienen la misma prioridad, se resuelven de izquierda a derecha.

| Prioridad | Siglas | Significado                         | Operadores       |
|----------|--------|-------------------------------------|------------------|
| 1°       | P      | Paréntesis                         | `( )`            |
| 2°       | E      | Exponentes (Potencias)             | `**`             |
| 3°       | M / D  | Multiplicación / División          | `*`, `/`, `//`, `%` |
| 4°       | A / S  | Adición / Sustracción              | `+`, `-`         |

Ejemplo: Promedio de calificaciones

```python
# Incorrecto: La división se hace antes que la suma
promedio_mal = 8 + 10 + 10/ 3  # Resultado: 13.0 (Error lógico)
#8 + 10 + 3.3333 = 21.333
print(promedio_mal)

# Correcto: Los paréntesis obligan a sumar primero
promedio_bien = (8 + 10 + 10) / 3 # Resultado: 9.0
#(18)/3
print(promedio_bien)
```

## IF

La sentencia if nos permite tomar desiciones en nuestro programa. Si la condición se cumple se ejecuta el bloque código

####if (La decisión simple)
evalúa una expresión booleana.

Sintaxis Paso a Paso:

1. Palabra reservada: if

2. Condición: Una pregunta (ej. edad >= 18).

3. Los dos puntos (:): OBLIGATORIOS. Significan "entonces hacer lo siguiente".

4. Indentación (Sangría): El espacio a la izquierda que indica qué código depende del if

```python
# SINTAXIS
condicion = False

if condicion:
  #bloque de codigo
  print("Condicion verdadera, Estamos dentro del if")
  print ( "Estamos en la clase de python")
else:
  print("Condicion falsa, Estamos dentro del else")
  print ( 20 + 8)

print(3*10)
print("Estamos afuera de la sentencia if")
```

####else
El else captura todo lo que no cumplió el if.

Regla de oro: El else nunca lleva una condición propia; es el "si no" por defecto.

```python
edad = int(input("Ingresa tu edad: "))

if edad >= 18 and edad<=30 :
    #condicion True
    print("Entrada permitida.")
else:
    print("Entrada denegada. No es mayor de 18 años y  menor a 30.")


print("Estamos fuera del if")
```

####elif (Múltiples opciones)

Python revisa el primer escalón (if). Si es falso, baja al siguiente (elif). Si todos son falsos, llega al suelo (else).

```python
puntos = 81


if puntos >= 100: #false
    print("Nivel: Super")
elif puntos >= 90:
    print("Nivel: Leyenda 🏆")
elif puntos >= 80:
    print("Nivel: Guerrero ⚔️")
    if puntos == 85:
      print("nivel: principiante 2")
    elif puntos ==81:
      print ("nivel: principiante 1")

else:
    print("Nivel: Novato 🌱")



print("Ya salimos del if")
```

####If Anidados (Decisiones dentro de decisiones)

"Imaginen que quieren entrar a una aplicación de banco. Primero, el sistema revisa si el Usuario existe. Solo si el usuario es correcto, entonces el sistema pasa a revisar si la Contraseña es correcta".

Si el usuario no existe: Ni siquiera te pregunta la contraseña.

```python
# Archivo: geometria.py

def calcular_area_circulo(radio):
    """Calcula el área de un círculo: pi * r^2"""
    pi = 3.14159
    return pi * (radio ** 2)

def calcular_area_rectangulo(base, altura):
    """Calcula el área de un rectángulo: base * altura"""
    return base * altura

def calcular_area_triangulo(base, altura):
    return (base * altura) / 2

def es_triangulo_valido(a, b, c):
    """Verifica si tres medidas pueden formar un triángulo (propiedad geométrica)"""
    return (a + b > c) and (a + c > b) and (b + c > a)
```

```python
# Archivo: main.py
import geometria  # Importamos nuestro archivo de herramientas

def menu():
    print("--- CALCULADORA GEOMÉTRICA ---")
    print("1. Área de un Rectángulo")
    print("2. Área de un Triángulo")
    print("3. Área de un Círculo")
    print("4. Salir")

def ejecutar():
    while True:
        menu()
        opcion = input("\nElige una opción (1-4): ")

        if opcion == "4":
            print("¡Hasta luego!")
            break

        try:
            if opcion == "1":
                b = float(input("Base: "))
                h = float(input("Altura: "))
                # Usamos el módulo: nombre_archivo.funcion()
                resultado = geometria.calcular_area_rectangulo(b, h)
                print(f"El área del rectángulo es: {resultado}")

            elif opcion == "2":
                b = float(input("Base: "))
                h = float(input("Altura: "))

                resultado = geometria.calcular_area_triangulo(b, h)
                print(f"El área del triángulo es: {resultado}")

            elif opcion == "3":
                r = float(input("Radio: "))
                resultado = geometria.calcular_area_circulo(r)
                print(f"El área del círculo es: {resultado}")

            else:
                print(" Opción no válida, intenta de nuevo.")

        except ValueError:
            print(" Error: Por favor ingresa solo números.")

        print("\n" + "-"*30)

if __name__ == "__main__":
    menu()
    ejecutar()
```

## MANEJO DE ARCHIVOS

El flujo básico de un archivo:

1. Abrir el archivo (open).

2. Trabajar con él (Leer o Escribir).

3. Cerrar el archivo (close o usar with).

2. Los Modos de Apertura
Es vital que entiendan que Python necesita saber qué vas a hacer con el archivo:

* 'r' (Read): Solo para leer. Si el archivo no existe, da error.

* 'w' (Write): Para escribir. Ojo: Si el archivo ya existe, borra todo lo anterior y empieza de cero.

* 'a' (Append): Para añadir. Escribe al final del archivo sin borrar lo que ya estaba.



| Modo | Nombre | Descripción | ¿Crea archivo? | ¿Borra contenido? |
|------|---------|-------------|----------------|------------------|
| `'r'` | Read | Abre el archivo para leer. | No (da error si no existe) | No |
| `'w'` | Write | Abre el archivo para escribir. | Sí | **SÍ (Sobrescribe)** |
| `'a'` | Append | Abre el archivo para añadir contenido al final. | Sí | No |
| `'r+'` | Read + Write | Permite lectura y escritura. | No | No |

3. La Sentencia with: El estándar profesional

archivo = open() y luego archivo.close() es "estilo antiguo".
El estándar moderno es el Context Manager (with).

¿Por qué usar with?

Cierra el archivo automáticamente incluso si ocurre un error en el código intermedio.

Es más limpio y evita fugas de memoria.

```python
# 1. Abrir el archivo
archivo = open("datos.txt", "w")

# 2. Realizar operaciones
archivo.write("Hola, este es un ejemplo de abrir y cerrar.")

# 3. Cerrar el archivo (CRUCIAL)
archivo.close()
```

```python
# No necesitas llamar a .close(), Python lo hace por ti al salir del bloque
with open("receta.txt", "w") as archivo:
    archivo.write("Ingredientes:\n- Harina\n- Huevos")
    # El archivo sigue abierto aquí dentro

# Aquí afuera, el archivo ya está cerrado
```

## PROGRAMACIÓN ORIENTADA A OBJETOS (OOP)

OOP significa Programación Orientada a Objetos .

Python es un lenguaje orientado a objetos, lo que permite estructurar el código utilizando clases y objetos para una mejor organización y reutilización.

Ventajas de la programación orientada a objetos (POO)

* Proporciona una estructura clara a los programas.
* Facilita el mantenimiento, la reutilización y la depuración del código.
* Ayuda a mantener tu código DRY ( No te repitas ).
* Te permite crear aplicaciones reutilizables con menos código.


Consejo: El principio DRY (Don't Repeat Yourself) implica evitar escribir el mismo código más de una vez. Traslada el código repetido a funciones o clases y reutilízalo.

¿Qué son las clases y los objetos?
Las clases y los objetos son los dos conceptos fundamentales de la programación orientada a objetos.

Una clase define cómo debe ser un objeto, y un objeto se crea en función de esa clase. Por ejemplo:

|Class	|Objects|
|-----|----------|
|Fruit|	Apple, Banana, Mango|
|Car	|Volvo, Audi, Toyota|


Cuando se crea un objeto a partir de una clase, este hereda todas las variables y funciones definidas dentro de esa clase.

```python
class Perro:
    def __init__(self, nombre, edad):
        self.nombre = nombre   # atributo
        self.edad = edad       # atributo

    def ladrar(self):          # método
        print("¡Guau guau!")

# Crear un objeto (instancia)
mi_perro = Perro("Firulais", 3)

print(mi_perro.nombre)  # Muestra: Firulais
mi_perro.ladrar()       # Muestra: ¡Guau guau!
```

```python
class Coche:
    # El constructor: define qué datos necesita el coche al "nacer"
    def __init__(self, marca, color):
        self.marca = marca    # Atributo
        self.color = color    # Atributo
        self.encendido = False

    # Un método (acción)
    def arrancar(self):
        self.encendido = True
        print(f"El {self.marca} de color {self.color} ha arrancado. ¡Brum brum!")
    def apagar(self):
        self.encendido = False
        print(f"El {self.marca} de color {self.color} ha apagado.")
#

# --- Usando el molde para crear objetos reales ---
mi_coche = Coche("Toyota", "Rojo")
coche_de_mi_amiga = Coche("Ferrari", "Negro")

mi_coche.arrancar()
coche_de_mi_amiga.arrancar()

mi_coche.apagar()
coche_de_mi_amiga.apagar()


```
